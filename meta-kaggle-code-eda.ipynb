# All the Notebook Json have these keys:
- 'cells'
    - The code and metadata for the notebook including:
        - `metadata` : about the code cell
        - `cell_type` : code or markdown?
        - `source` : actual code or text for the cell
        for "code"
            - `execution_count` count of times it was run?
            - `outputs` : not included
- 'metadata'
    - `kernelspec` - python or R and version info
    - `language_info` - more detailed info about the language
- `nbformat` - nbformat version number
- `nbformat_minor`

# Look into the Code!

```python
def load_ipynb(file):
    with open(file, 'r') as f:
        return json.load(f)
    

def extract_code(data):
    codes = []
    for c in data['cells']:
        if c['cell_type'] == 'code':
            codes.append(c['source'])
    return codes

def extract_markdown(data):
    mds = []
    for c in data['cells']:
        if c['cell_type'] == 'markdown':
            mds.append(c['source'])
    return mds
```

# Load all the code from the ENTIRE 1M (potentially) folder

```python
len(ipynb_files)
```

```python
results = {}
for file in tqdm(ipynb_files):
    file_number = file[:-6].split('/')[-1]
    data = load_ipynb(file)
    codes = extract_code(data)
    results[int(file_number)] = codes
```

```python
all_codes = []
for file_number, codes in results.items():
    all_codes += codes
```
